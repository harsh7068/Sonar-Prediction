```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
data=pd.read_csv('sonar.csv',header=None)
data
```

```python
data.shape
```

```python
data.describe()
```

```python
data[60].value_counts()
```

```python
data.groupby(60).mean()
```

# separating data and labels

```python
x=data.drop(columns=60, axis=1)
y=data[60]
```

```python
print(x)
print(y)
```

```python
#training and testing data
train_x,test_x,train_y,test_y=train_test_split(x,y,test_size=0.1,stratify=y,random_state=0)
```

```python
train_x
```

```python
test_x
```

```python
train_y
```

```python
test_y
```

```python
print(train_x.shape)
print(train_y.shape)
print(test_x.shape)
print(test_y.shape)
```

```python
model=LogisticRegression()
```

# model training

```python
model.fit(train_x,train_y)
```

# Model Evaluation

```python
x_pred=model.predict(train_x)
x_pred
```

```python
x_acu=accuracy_score(x_pred,train_y)
print('accuracy on data:',x_acu)
```

```python
x_pred_test=model.predict(test_x)
x_pred_test
```

```python
x_acu_test=accuracy_score(x_pred_test,test_y)
print('accuracy on data:',x_acu_test)
```
